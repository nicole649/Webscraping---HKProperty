```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('TKS.csv')
df.head()
```

```python
df = df.iloc[:,0:12]
df.drop('(Saleable).1', axis=1, inplace=True)
```

```python
column=['buliding_name','floor','flat','area_saleable','area_gross','price(m)','price_ft2_sale','price_ft2Gross','trans_date','change','last_transaction']
df.columns = column  #update columns' names
df = df.drop_duplicates()  #remove duplicate records
df
```

```python
df[df.flat == 'flat']
```

```python
ndf= df[df.flat != 'flat']
```

```python
ndf.info()
```

```python
ndf[['estate','building']] = ndf['buliding_name'].str.split('\n', 1, expand=True)
```

```python
ndf = ndf.reset_index()
```

```python
ndf[ndf.trans_date == '9,177']
```

```python
ndf.loc[9118,'trans_date'] = '1997-11-26'
ndf.loc[9118,'price_ft2Gross'] = '9,177'
ndf.loc[9118,'change'] = np.nan
```

```python
ndf.iloc[9118]
```

```python
ndf['area_saleable'] = ndf['area_saleable'].str.replace(',','')
ndf['area_gross'] = ndf['area_gross'].str.replace(',','')
ndf['price(m)'] = ndf['price(m)'].str.replace(',','')
ndf['price_ft2_sale'] = ndf['price_ft2_sale'].str.replace(',','')
ndf['price_ft2Gross'] = ndf['price_ft2Gross'].str.replace(',','')
ndf['area_gross'] = ndf['area_gross'].replace('-',np.NAN)
ndf['price(m)'] = ndf['price(m)'].replace('-',np.NAN)
ndf['price_ft2_sale'] = ndf['price_ft2_sale'].replace('-',np.NAN)
ndf['price_ft2Gross'] = ndf['price_ft2Gross'].replace('-',np.NAN)
```

```python
ndf[ndf.area_saleable == '*']
ndf.loc[9118,'area_saleable'] = np.nan
```

```python
ndf.head()
ndf = ndf.astype({"area_saleable": float,
                   "area_gross": float,
                   "price(m)": float,
                   "price_ft2_sale": float,
                   "price_ft2Gross":float})
ndf.area_saleable = ndf.area_saleable.astype('Int64')
ndf['price(m)'] = ndf['price(m)']/100  # the unit should be in millions instead of 10 thousands
```

```python
ndf.trans_date = pd.to_datetime(ndf.trans_date,format='%Y/%m/%d')
```

```python
ndf['year_Q'] = ndf['trans_date'].dt.to_period("Q")
```

```python
ndf.drop(columns=['index','buliding_name','area_gross','price_ft2Gross','change','last_transaction','trans_date'],inplace=True)
```

```python
ndf.dropna(how='any',inplace =True)
```
